```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
df_IN = pd.read_pickle("IN_training_output.pkl")
df_LDMX_IN = pd.read_pickle("LDMX_IN_training_output.pkl")
```

```python
#plt.plot(np.arange(len(df_IN))+1, df_IN['train_loss'], label='LDMX IN Train Loss')
#plt.plot(np.arange(len(df_LDMX_IN))+1, df_LDMX_IN['train_loss'], label='IN Train Loss')
plt.plot(np.arange(len(df_LDMX_IN))+1, df_LDMX_IN['val_loss'], color='blue', label='LDMX IN Val Loss')
plt.plot(np.arange(len(df_IN))+1, df_IN['val_loss'], color='orange', label='IN Val Loss')



plt.axvline(x=np.argmin(df_LDMX_IN['val_loss'])+1, linestyle='--', color='blue', label='LDMX IN Min Val Loss Epoch')
plt.axvline(x=np.argmin(df_IN['val_loss'])+1, linestyle='--', color='orange', label='IN Min Val Loss Epoch')

plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.yscale('log')

```

```python
# Get the test accuracy at the epoch with the lowest validation loss
best_epoch_LDMX_IN = np.argmin(df_LDMX_IN['val_loss'])
best_test_acc_LDMX_IN = df_LDMX_IN['test_acc'][best_epoch_LDMX_IN]

best_epoch_IN = np.argmin(df_IN['val_loss'])
best_test_acc_IN = df_IN['test_acc'][best_epoch_IN]

print(f"LDMX IN Best Test Accuracy at Epoch {best_epoch_LDMX_IN + 1}: {best_test_acc_LDMX_IN:.6f}")
print(f"IN Best Test Accuracy at Epoch {best_epoch_IN + 1}: {best_test_acc_IN:.6f}")

```
